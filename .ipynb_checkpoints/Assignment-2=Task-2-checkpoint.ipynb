## 1Ô∏è‚É£ Unique Words in a Sentence

### Problem Statement
Given a sentence, identify all unique words and display:
1. The count of unique words
2. The unique words themselves

### Approach
- Split the sentence into individual words using split()
- Convert the list into a set to remove duplicates
- Print the set and its length

### Real-World Applications
- Text analytics
- NLP preprocessing
- Search engine indexing
- Chat analysis systems


```python
sentence = "python is easy and python is powerful"

# Split sentence into words
words = sentence.split()

# Convert list to set to remove duplicates
unique_words = set(words)

print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

## 2Ô∏è‚É£ Highest Salary from Employee Data

### Problem Statement
Store employee names and their salaries, and determine which employee has the highest salary.

### Approach
- Store data in a dictionary
- Use max() with key=employees.get to find highest salary
- Print employee name and salary

### Real-World Applications
- HR management systems
- Payroll analysis


```python
employees = {
    "Anshika": 75000,
    "Deepti": 68000,
    "Neha": 72000
}

highest_employee = max(employees, key=employees.get)
highest_salary = employees[highest_employee]

print(f"Highest Salary: {highest_employee} - {highest_salary}")
```

## 3Ô∏è‚É£ Find Maximum and Minimum Values

### Problem Statement
Find the maximum and minimum values from a list of numbers.

### Approach
- Initialize max and min with first element
- Use loop to compare each value
- Update max and min accordingly

### Real-World Applications
- Sales data analysis
- Temperature monitoring systems
- Performance metric evaluation


```python
numbers = [45, 22, 89, 10, 66]

maximum = numbers[0]
minimum = numbers[0]

for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

## 4Ô∏è‚É£ Count Products Above a Price Threshold

### Problem Statement
Count how many products have a price greater than 1000.

### Approach
- Loop through price list
- Compare each price with 1000
- Increment counter if condition satisfied

### Real-World Applications
- E-commerce product filtering
- Budget analysis


```python
prices  = [ 450 , 1200 , 899 , 1500 , 300]

count = 0

for i in prices:
    if i > 1000:
        count += 1

print("Products above 1000  = " , count)
```

## 5Ô∏è‚É£ Calculate Attendance Percentage

### Problem Statement
Calculate attendance percentage from a list containing "P" (Present) and "A" (Absent).

### Approach
- Count number of "P"
- Divide by total days
- Multiply by 100

### Real-World Applications
- Internship tracking systems
- HR attendance management systems


```python
attendance = ["P", "P", "A", "P", "P"]

present_count = attendance.count("P")
total_days = len(attendance)

percentage = (present_count / total_days) * 100

print("Attendance Percentage:", percentage)
```

## 7Ô∏è‚É£ Count Character Frequency

### Problem Statement
Count how many times each character appears in a string.

### Approach
- Use dictionary
- Increment count for each character
- Print dictionary

### Real-World Applications
- Text analysis
- Password strength validation


```python
text = "Python"

char_count = {}

for char in text:
    if char in char_count:
        char_count[char] += 1
    else:
        char_count[char] = 1

print(char_count)
```

## 9Ô∏è‚É£ Check if a Key Exists in a Dictionary

### Problem Statement
Check whether a specific key exists in a dictionary.

### Approach
- Use 'in' keyword
- Print appropriate message

### Real-World Applications
- User validation systems
- Configuration verification


```python
employees = { "Anshika": 75000, "Neha": 68000 }

key_to_check = "Neha"

if key_to_check in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

## üîü Calculate Average Marks

### Problem Statement
Calculate the average marks from a list of student marks.

### Approach
