## 1️⃣ User Login Check

### Problem Statement
Given a username and password, check whether login is successful.

### Real-World Application
Used in authentication systems to validate user credentials.


```python
# Given credentials
username = "admin"
password = "1234"

# User input simulation (can be replaced with input() if needed)
entered_username = "admin"
entered_password = "1234"

# Login check
if entered_username == username and entered_password == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

## 2️⃣ Pass / Fail Analyzer

### Problem Statement
Analyze student results and determine pass and fail counts.

### Real-World Application
Used in academic evaluation systems.


```python
marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("Total Students:", len(marks))
print("Pass Students:", pass_count)
print("Fail Students:", fail_count)
```

## 3️⃣ Simple Data Cleaner

### Problem Statement
Clean and standardize user names.

### Real-World Application
Data preprocessing before analysis.


```python
names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

for name in names:
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)

print("Cleaned Names:", cleaned_names)
```

## 4️⃣ Message Length Analyzer

### Problem Statement
Analyze message sizes and flag long messages.

### Real-World Application
Text filtering and validation systems.


```python
messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages:
    length = len(message)
    print(f"Message: '{message}' | Length: {length}")
    
    if length > 10:
        print("⚠️ This message is longer than 10 characters")
```

## 5️⃣ Error Message Detector

### Problem Statement
Detect error messages from system logs.

### Real-World Application
Monitoring and log analysis systems.


```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = logs.count("ERROR")

print("Total ERROR Messages:", error_count)
```
