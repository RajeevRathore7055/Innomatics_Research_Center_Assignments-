# Problem 1: Smart Light Controller

## Objective
Create a class that controls a smart light system which can be turned ON or OFF and display its current status.

## OOPS Concept Used
- Class and Object
- Instance Variables
- Methods

## Real-World Application
- Smart home automation systems
- IoT-based lighting control

```python
class SmartLight:
    
    def __init__(self, light_name):
        self.light_name = light_name
        self.status = "OFF"
    
    def turn_on(self):
        self.status = "ON"
    
    def turn_off(self):
        self.status = "OFF"
    
    def display_status(self):
        print(f"{self.light_name} is {self.status}")


# Sample Input
light = SmartLight("Bedroom Light")
light.turn_on()
light.display_status()
```

# Problem 2: Employee ID Card System

## Objective
Store employee details and display formatted ID card information.

## OOPS Concept Used
- Constructor (__init__)
- Instance attributes
- Display method

## Real-World Application
- Corporate HR systems
- Employee management software

```python
class Employee:
    
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department
    
    def display_id_card(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


# Sample Input
emp = Employee("Rahul", "EMP102", "IT")
emp.display_id_card()
```

# Problem 3: Mobile Contact Record

## Objective
Store and display contact details using class structure.

## OOPS Concept Used
- Object creation
- Data encapsulation

## Real-World Application
- Mobile contact management apps

```python
class Contact:

    def __init__(self , name , phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("Contact saved")
        print("Name" , self.name)
        print("Phone" , self.phone)

# Sample Input
contact = Contact("Anita", 9876543210)
contact.display_contact()
```

# Problem 1: Product Price Tag Generator

## Objective
Generate formatted product price tag using class.

## OOPS Concept Used
- Attribute storage
- Method for formatted output

## Real-World Application
- Retail billing systems
- E-commerce platforms

```python
class Product:
    
    def __init__(self, name, price):
        self.name = name
        self.price = price
    
    def price_tag(self):
        print("Product:", self.name)
        print("Price: ₹{}".format(self.price))


# Sample Input
product = Product("Headphones", 2499)
product.price_tag()
```

# Problem 5: Movie Rating Display System

## Objective
Store movie name and rating and display formatted rating information.

## OOPS Concept Used
- Class and methods
- Instance attributes

## Real-World Application
- Streaming platforms
- Movie review systems

```python
class Movie:
    
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating
    
    def display_movie(self):
        print("Movie:", self.name)
        print("Rating:", self.rating, "/ 5")


# Sample Input
movie = Movie("Inception", 4.8)
movie.display_movie()
```

# Problem 6: Delivery Address Manager

## Objective
Store customer and delivery address details and print formatted output.

## OOPS Concept Used
- Constructor
- Instance variables
- Display method

## Real-World Application
- Courier systems
- Logistics platforms

```python
class Delivery:

    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address
    
    def display_details(self):
        print("Delivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)


# Sample Input
delivery = Delivery("Suman", "Hyderabad")
delivery.display_details()
```
