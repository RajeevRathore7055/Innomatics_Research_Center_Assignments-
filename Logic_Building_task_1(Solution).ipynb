##2. **Pass / Fail Analyzer**

**Problem Statement:**
Analyze student results.
marks = [45, 78, 90, 33, 60]

**Requirements:**

● A student passes if marks ≥ 50

● Count the total number of pass students

● Count the total number of fail students

● Print the final result clearly
Real-World Application: Academic evaluation systems

```python
def Pass_Fail_Analyzer(marks):
  pass_count = 0
  fail_count = 0

  for mark in marks:
    if mark>=50:
      pass_count += 1
    else:
      fail_count +=1

  print(f"Total Number of student passed in the test: {pass_count}")
  print(f"Total Number of student failed in the test: {fail_count}")


if __name__ == "__main__" :

   marks = [45,78,90,33,60]
   print("Analyzing student result:")
   Pass_Fail_Analyzer(marks)


```

##5. **Error Message Detector**

**Problem Statement:**

Detect error messages from system logs.

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

**Requirements:**

- Count the number of "ERROR" entries

- Print the total error count

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

count = 0

for log in logs:
    if log == "ERROR":
        count = count + 1

print(f"The number of ERROR in the list is {count}")
```
